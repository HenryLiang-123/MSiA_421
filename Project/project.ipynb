```python
import pandas as pd
import numpy as np

dat = pd.read_csv('lineups_with_clusters.csv')
dat.head(5)
```

```python
dat_selected = dat.loc[:, [' NET_RATING', ' MIN', 'p1_cluster', 'p2_cluster', 'p3_cluster', 'p4_cluster', 'p5_cluster']]
```

```python
dat_selected
```

```python
for i in range(5):
    print(dat_selected[f'p{i+1}_cluster'].unique())
```

```python
3 + np.nan
```

```python
total_dummies = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
dummies_no4 = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
dummies_4 = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
for i in range(3):
    dummies_no4 += pd.get_dummies(dat_selected[f'p{i+1}_cluster'], dummy_na=True)

for i in range(2):
    dummies_4 += pd.get_dummies(dat_selected[f'p{i+4}_cluster'], dummy_na=True)

dummies_no4.fillna(0, inplace=True)
total_dummies = dummies_no4 + dummies_4
```

```python
total_dummies
```

```python
total_dummies.isnull().sum()
```

```python
dat_final = pd.concat([dat_selected.loc[:, [' NET_RATING', ' MIN']], total_dummies], axis=1)
```

```python
dat_final.columns
```

```python
dat_final.rename(columns={' NET_RATING': 'net_rating', " MIN": 'minutes', 0: 'cluster_0', 1: 'cluster_1', 2: 'cluster_2', 3: 'cluster_3', 4: 'cluster_4', 5: 'cluster_5', 6: 'cluster_6', 7: 'cluster_7',}, inplace=True)
dat_final
```

```python
dat_final.describe()
```

```python
min_50 = dat_final.loc[dat_final['minutes'] >= 50, :]
```

```python
len(min_50)
```

## Start with Linear Regression

```python
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split

X = dat_final.drop(labels=['net_rating'], axis=1)
y = dat_final['net_rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
formula = 'net_rating ~ cluster_0 + cluster_1 + cluster_2 + cluster_3 + cluster_4 + cluster_5 + cluster_6 + cluster_7'
m1 = smf.ols(formula=formula, data = pd.concat([X_train, y_train], axis = 1))
m1_fit = m1.fit()
m1_fit.summary()
```

```python
# LR for >50 minutes

min_50 = dat_final.loc[dat_final['minutes'] >= 50, :]
X = min_50.drop(labels=['net_rating'], axis=1)
y = min_50['net_rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
formula = 'net_rating ~ cluster_0 + cluster_1 + cluster_2 + cluster_3 + cluster_4 + cluster_5 + cluster_6 + cluster_7'
m1 = smf.ols(formula=formula, data = pd.concat([X_train, y_train], axis = 1))
m1_fit = m1.fit()
m1_fit.summary()
```

```python
import statistics as stat

pred = m1_fit.predict(X_test)
mse = np.mean((pred - y_test)**2)
r2 = 1 - mse / stat.variance(y_test)
r2
```

```python
pred = m1_fit.predict(X_test)
mae = np.mean(abs(pred - y_test))
mae
```

Pretty poor, so not going to waste time doing testing or cross validation

# Try xgboost

```python
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score

X = min_50.drop(labels=['net_rating'], axis=1)
y = min_50['net_rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
m2 = XGBRegressor()
m2.fit(X_train, y_train)

scores_mae = cross_val_score(m2, X_train, y_train,scoring='neg_mean_absolute_error', cv= 10)
-scores_mae.mean()
```

```python
scores_r2 = cross_val_score(m2, X_train, y_train,scoring='r2', cv= 10)
scores_r2.mean()
```

## Try Random Forest

Always the best

```python
from sklearn.ensemble import RandomForestRegressor

# Build basic model to see if it is worth doing grid search to find best params
X = min_50.drop(labels=['net_rating'], axis=1)
y = min_50['net_rating']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
rf = RandomForestRegressor(n_estimators = 30, random_state = 42, oob_score=True)
rf.fit(X_train, y_train)
```

```python
len(X_train)
```

```python
rf.score(X_train, y_train)
```

```python
rf.oob_score_
```

```python
pred = rf.predict(X_test)
mae = np.mean(abs(pred - y_test))
mae
```

```python
import statistics as stat

mse = np.mean((pred - y_test)**2)
r2 = 1 - mse / stat.variance(y_test)
r2
```
