```python
import pandas as pd
import numpy as np

dat = pd.read_csv('lineups_with_clusters.csv')
dat.head(5)
```

```python
dat_selected = dat.loc[:, [' NET_RATING', 'p1_cluster', 'p2_cluster', 'p3_cluster', 'p4_cluster', 'p5_cluster']]
```

```python
dat_selected
```

```python
for i in range(5):
    print(dat_selected[f'p{i+1}_cluster'].unique())
```

```python
3 + np.nan
```

```python
total_dummies = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
dummies_no4 = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
dummies_4 = pd.DataFrame(np.zeros((dat_selected.shape[0], 8)))
for i in range(3):
    dummies_no4 += pd.get_dummies(dat_selected[f'p{i+1}_cluster'], dummy_na=True)

for i in range(2):
    dummies_4 += pd.get_dummies(dat_selected[f'p{i+4}_cluster'], dummy_na=True)

dummies_no4.fillna(0, inplace=True)
total_dummies = dummies_no4 + dummies_4
```

```python
total_dummies
```

```python
total_dummies.isnull().sum()
```

```python
dat_final = pd.concat([dat_selected[' NET_RATING'], total_dummies], axis=1)
```

```python
dat_final.columns
```

```python
dat_final.rename(columns={' NET_RATING': 'net_rating', 0: 'cluster_0', 1: 'cluster_1', 2: 'cluster_2', 3: 'cluster_3', 4: 'cluster_4', 5: 'cluster_5', 6: 'cluster_6', 7: 'cluster_7',}, inplace=True)
dat_final
```

```python
dat_final.describe()
```

## Start with Linear Regression

```python
import statsmodels.formula.api as smf

formula = 'net_rating ~ cluster_0 + cluster_1 + cluster_2 + cluster_3 + cluster_4 + cluster_5 + cluster_6 + cluster_7'
m1 = smf.ols(formula=formula, data = dat_final)
m1_fit = m1.fit()
m1_fit.summary()
```
